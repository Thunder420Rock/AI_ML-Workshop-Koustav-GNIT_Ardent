```python
#Step 1:import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Optional Display setting
pd.set_option("display.max_columns",None)
```

```python
#step 2 : Data Collection
#Creating a sa,ple student data set

data ={
    "Student_ID": [101,102,103,104,105,106,107,108],
    "Name":["Rahul","Madhumita","Sara","Amit","Priya","Jhon","Riya","Arjun"],
    "Gender":["Male","Female","Female","male","Female","male","Female","male"],
    "Math":[85,92,78,np.nan,88,76,97,78],
    "Science":[90,89,np.nan,72,85,67,89,90],
    "English":[88,94,75,74,np.nan,78,65,66],
    "Attendence":[92,96,85,78,90,69,88,89]
}
df = pd.DataFrame(data)
print("Original Data set")
print(df)
```

```python
#Step 3:Basic Information
print("\n============== DATASET INFO ===============")
print(df.info())
print("\n============== 1st FIVE ROWS ==============")
print(df.head())
print("\n======== Last 5 Rows ===========")
print(df.tail)
print("\n========== SHAPE OF DATA SET =======")
print("Rows,Column:",df.shape)

print("\n =================== SHAPE OF DATASET =========")
print(df.columns)
```

```python
#step 4:Check missing Values
print("\n========== MISSING VALUEs =============")
print(df.isnull().sum())

```

```python
#Step 5: Data Preprocessing
# Filling misiig numeri values with column mean

df["Math"].fillna(df["Math"].mean(), inplace=True)
df["Science"].fillna(df["Science"].mean(), inplace=True)
df["English"].fillna(df["English"].mean(), inplace=True)
df["Attendence"].fillna(df["Attendence"].mean(), inplace=True)

print("\n ======== DATASET AFTER FILLING MISSING VAlUES ========= ")
print(df)
print("\n ======== MISSING VALUES AFTER CLEANING ========= ")
print(df.isnull().sum())
```

```python
#Step 6:ADD new Columns
# ==================================
# Total and Average marks

df["Total"] = df["Math"] + df["Science"] + df["English"]
df["Average"] = df["Total"]/3

#Grade Calculation
def assign_grade(avg):
  if avg >= 90:
    grade = "A+"
  elif avg >= 80:
    grade = "A"
  elif avg >= 70:
    grade = "B"
  elif avg >= 60:
    grade = "C"
  elif avg >= 50:
    grade = "D"
  else:
    grade = "F"

df["Grade"] = df["Average"].apply(assign_grade)

print("\n ======= DATASET With Total,Average, Grade ==============")
print(df)
```
